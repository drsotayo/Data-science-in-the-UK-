### Functions

```python
def grab_subset_of_data(original_df, question_of_interest):
    """
    This function grabs gets a a subset of the data from the original dataframe based on pattern matching. 
    """
    subset_of_columns = [col for col in original_df.columns if question_of_interest in col]
    mapping_dict = {col: col.split('-')[-1].lstrip() for col in subset_of_columns}
    new_df = original_df[subset_of_columns].rename(columns=mapping_dict)
    new_df.dropna(how='all', subset=mapping_dict.values(), inplace=True)
    return new_df
```

```python
def select_column(dataframe,x_axis_title,new_col_name): 
    """
    This function selects and renames a column from the specified dataframe and drops null values.
    """
    interested_column = dataframe[[x_axis_title]].dropna()
    interested_column = interested_column.rename(columns={x_axis_title:new_col_name})
    return interested_column
```

```python
def bar_chart_multiple_choice_multiple_selection_seaborn(df,title,orientation='v', marker_color=british_philippine_red): 
    """
    This code gets a subset of questions and plots a bar, it requires the main dataset and title of the chart.    
        
    """
    df_new = df.copy()
    counts = df_new[df_new.columns[:]].count().sort_values(ascending=False)
    percentages = counts*100/len(df_new)
    percentages.index.name = ''
    fig = plt.figure(figsize=(10, 7))  # Set figure size
    if orientation is 'v':
        ax = sns.barplot(x=percentages.index, y=percentages.values, color=marker_color)
        ax.set_title(title+' in the UK in 2022')
        ax.set_ylabel('Percentage of survey respondents (%)')
        ax.set_yticklabels(['{:,.0f}%'.format(y) for y in ax.get_yticks()])
        #ax.set_xticks([])  
        for p in ax.patches:
            ax.text(p.get_x() + p.get_width()/2, p.get_height(), '{:.0f}%'.format(p.get_height()), ha="center", va="bottom")
    else:
        ax = sns.barplot(x=percentages.values, y=percentages.index, color=marker_color)
        ax.set_title(title+' in the UK in 2022')
        ax.set_xlabel('Percentage of survey respondents (%)')
        ax.set_xticklabels(['{:,.0f}%'.format(x) for x in ax.get_xticks()])
        ax.set_xticks([]) 
        for p in ax.patches:
            ax.text(p.get_width(), p.get_y() + p.get_height()/2, '{:.0f}%'.format(p.get_width()), ha="left", va="center")
    sns.despine(left=True, bottom=True)
```

```python
def bar_chart_with_seaborn(y_series, x_series, chart_title,
                           marker_color=british_catalina_blue, 
                           orientation="h", vertical_line ="n"):
    fig = plt.figure(figsize=(10, 7))  # Set figure size
    if orientation == "h":
        ax = sns.barplot(y=list(reversed(y_series)), x=list(reversed(x_series)), color=marker_color)
        #ax.set_title(chart_title)
        ax.set_title(chart_title+' in 2022')
        ax.set_xlabel('British Pound, GBP (£)')
        ax.set_xticklabels(['£{:,.0f}k'.format(round(x/1000)) for x in ax.get_xticks()])
        ax.set_xticks([])
        ax.set_ylabel('')
        for p in ax.patches:
            ax.text(p.get_width(), p.get_y() + p.get_height()/2, '{:,.0f}k'.format(round(p.get_width()/1000)), 
                    ha="left", va="center")
        if vertical_line == "y" and orientation == "h":
            ax.axvline(x=uk_median_salary, linestyle="dotted", 
                       color=british_philippine_red, label="Median Salary for Data Professionals in the UK")
            ax.legend(loc="lower right")
    else:
        ax = sns.barplot(x=list(reversed(y_series)), y=list(reversed(x_series)), color=british_catalina_blue)
        #ax.set_title(chart_title)
        ax.set_title(chart_title+' in 2022')
        ax.set_ylabel('British Pound, GBP (£)')
        ax.set_yticklabels(['£{:,.0f}k'.format(round(y/1000)) for y in ax.get_yticks()])
        #ax.set_yticklabels('')
        #ax.set_yticks([])
        for p in ax.patches:
            ax.text(p.get_x() + p.get_width()/2, p.get_height(), '{:,.0f}k'.format(round(p.get_height()/1000)), 
                    ha="center", va="bottom")
    sns.despine(left=True, bottom=True)
```

```python
def bar_chart_multiple_choice(response_counts,title,y_axis_title,orientation='h',num_choices=15): 
    """
    This code gets a subset of questions, counts the number of responses and plots a bar.
    It requires the main dataset and title of the chart.      
    This code was adapted from https://www.kaggle.com/code/paultimothymooney/kaggle-survey-2022-all-results 
    """
    response_counts_series = pd.Series(response_counts)
    response_counts_series.index.name = ''
    fig = px.bar(response_counts_series,
             labels={"": '',"value": y_axis_title},
             text=response_counts_series.values,
             orientation=orientation,)
    fig.update_layout(showlegend=False,
                      title={'text': title+' in 2022',
                             'y':0.95,
                             'x':0.5,},
                      xaxis_range=[0,max(response_counts_series[0:num_choices]*1.2)])
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside',marker_color=british_catalina_blue) 
    fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide', plot_bgcolor = "white")
    fig.show()
```

```python
def get_interested_column_with_salary(interested_column, salary_df_2022, group_by_column):
    interested_column_with_salary = interested_column.groupby(group_by_column).filter(lambda x: len(x)>10)
    #filters out response that are 10 or less due to stastistical insignificance.
    interested_column_with_salary = interested_column_with_salary.merge(salary_df_2022, left_index=True, 
                                                                        right_index=True, how="left")
    interested_column_with_salary = interested_column_with_salary.groupby(group_by_column).median().round(2).dropna()
    interested_column_with_salary["gbp"] = round(interested_column_with_salary["What is your current yearly compensation (approximate $USD)?"]
                                   *usd_to_gbp_rate)
    interested_column_with_salary = interested_column_with_salary.sort_values(by="gbp", ascending=True)
    return interested_column_with_salary
```

```python
def get_job_with_skills(job_role, programming_language_df_2022, group_by_column):
    job_role_with_skill = job_role.merge(programming_language_df_2022, left_index=True,
                             right_index=True, how="left")
    job_role_with_skill = job_role_with_skill.set_index(group_by_column)
    job_role_with_skill = job_role_with_skill.notnull().astype("int")
    job_role_with_skill = job_role_with_skill.groupby(group_by_column).mean().round(2)
    return job_role_with_skill
    
```

```python
def plot_heatmap(job_role_with_skill):
    plt.figure(figsize = (15,10))
    sns.heatmap(job_role_with_skill, cmap = 'Blues', annot = True, cbar=False,fmt=".0%", linewidths=5)
    plt.ylabel("")
    plt.show()
    
```

## <span style='color:blue'> What's the pay like? </span>

<div class="alert alert-block alert-info">
<b>Assumptions made</b> 
</div>

- As the analysis is focussed on the UK and the salaries were reported in USD, the average 2022 exchange rate (USD to GBP) of ~ 0.81 was used for the salaries to convert it to British Pounds (GBP, £) [[Exchange Rates UK, 2022]](https://www.exchangerates.org.uk/USD-GBP-spot-exchange-rates-history-2022.html). 

- For this salary, the median is a better metric for getting a 'typical' salary compared to the mean due to the sample size (only UK survey respondents) and the presence of outliers.


For many, salary is a strong motivation and based on the survey responses, Data Science Professionals in the UK have a **median salary of about £61,000**, which is almost twice the median salary for all full-time employees in the UK in 2022 [[Statista, 2022]](https://www.statista.com/statistics/1002964/average-full-time-annual-earnings-in-the-uk/). However, this varies by job title, sector and years of experience.

On average, **Managers** in the UK earn the highest, which aligns well with years of experience and leadership responsibilities. Although sometimes used interchangeably by organisations and recruiters, **Data Scientists** earn an average of about **£61,000**, which is about **24% greater** than the average salary for a **Data Analyst (£49,000)** in the UK. *It is worth noting that job titles with less than 10 survey respondents (e.g. Data Engineers) in the UK were dropped as they were deemed statistically insignificant to make a reasonable conclusion.*

In addition, Glassdoor is a useful reference for comparing salaries for different roles and companies. It provides anonymous reviews and insights from former and current employees. For example, compared to the survey data, Glassdoor states that the average salaries for a Data Scientist and Data Analyst are £50,000 and £35,000, respectively, in the UK [[Glassdoor, 2022]](https://www.glassdoor.co.uk/Salaries/data-scientist-salary-SRCH_KO0,14.htm).

Another useful reference is Reed. Reed is a popular recruitment agency in the UK. Based on the data roles advertised on Reed, the average salaries for a Data Scientist and Data Analyst are £59,000 and £41,000, respectively in the UK [[Reed, 2022]](https://www.reed.co.uk/average-salary). 

Therefore, one of the key findings from the 2022 Kaggle Survey, Glassdoor and Reed is that **Data Scientists** typically earn about **24 - 44%** higher than **Data Analysts** in the UK. This means aspiring Data professionals in the UK should use tools like Glassdoor and Reed to research salaries during salary negotiations to make a more informed decision. Furthermore, organisations should know the salary ranges for different data roles to stay competitive and attract the best talents. Course developers can also inform their learners of the full range of salaries for different roles so that they can have realistic expectations.


```python
question_of_interest = 'What is your current yearly compensation (approximate $USD)?'
salary_df_2022 = grab_subset_of_data(uk, question_of_interest)
salary_df_2022 = salary_df_2022[question_of_interest].apply(lambda x: salary_map[x]) #uses the average value of salary
uk_median_salary = salary_df_2022.median()*usd_to_gbp_rate

question_of_interest = 'Select the title most similar to your current role (or most recent title if retired): - Selected Choice'
new_col_name = "job_role"
group_by_column = "job_role"
job_role = select_column(uk,question_of_interest,new_col_name)

job_role_with_salary = get_interested_column_with_salary(job_role, salary_df_2022, group_by_column)
y_series = job_role_with_salary.index
x_series = job_role_with_salary.gbp
chart_title = "Median salary for different Data Professionals in the UK"
bar_chart_with_seaborn(y_series, x_series, chart_title, vertical_line="y")
```
